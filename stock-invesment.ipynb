```python
data2 = pd.read_excel('中证全指.xls')
data2 = data2[:-1]
```

```python
import numpy as np
values2 = data2.values
temp1 = (values2[1:,2] - values2[1:,3])/values2[:-1,4]
temp2 = abs(values2[1:,2] - values2[:-1,4])/values2[:-1,4]
temp3 = abs(values2[1:,3] - values2[:-1,4])/values2[:-1,4]
TR = np.max(np.vstack((temp1,temp2,temp3)),axis = 0)
```

```python
def moving_average(a, n) :
    #temp = a[:n]
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    #return np.hstack((temp,ret[n - 1:] / n))
    return ret[n - 1:] / n
ATR = moving_average(TR,60)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure(dpi=200)
plt.title('ATR-TIME')
plt.xlabel('day')
plt.ylabel('ATR')
plt.plot(ATR)
plt.savefig('ATR.jpg')
```

```python
#计算每一天的下跌和反弹阈值（根据ATR调整过后的）
D = np.ones(1213-60) * (-0.05)
U = np.ones(1213-60) * (0.005)
for i in range(1,1213-60):
    if ATR[i] > 0.02:
        D[i] *= np.sqrt(ATR[i]/0.02)
        U[i] *= np.sqrt(ATR[i]/0.02)
    if ATR[i] < 0.01:
        D[i] *= np.sqrt(ATR[i]/0.01)
        U[i] *= np.sqrt(ATR[i]/0.01)        
```

```python
plt.figure(figsize = (12,6),dpi=200)

plt.subplot(121)
plt.plot(D)
plt.xlabel('day')
plt.ylabel('value')
plt.title('D')
plt.subplot(122)
plt.plot(U)
plt.xlabel('day')
plt.ylabel('value')
plt.title('U')
plt.savefig('DU.jpg')
```

### 计算全部的信号日

```python
#开始寻找信号日
day_list = []
fantan_qujian = []
fantan_qujian2 = []
for i in range(60,1213):
    for j in range(0,i):
        #i是当前日，j是区间开始的头
        if values2[i,4] - values2[i-1,4] < 0:
            #说明不在反弹的过程中，肯定不是我们要的
            break
        
        start_ = i-1
        while start_ > 0:
            if values2[start_,4] - values2[start_-1,4] > 0:
                start_ -= 1
            else:
                break
        if (values2[i,4] - values2[start_,4])/values2[start_,4] < U[i-60]:
            #触底反弹幅度不够，直接测试下一个i
            break
            
        max_v = np.max(values2[j:i,4])
        min_v = np.min(values2[j:i,4])
        if (min_v - max_v)/max_v > D[i-60]:
            #这段区间的下跌深度不够
            continue
            #测试下一个j
            
        m = j
        flag = 0
        flag2 = 0
        while m < i:
            if values2[m,4] > values2[m-1,4]:
                flag2 = 1
                start = m-1
                while (values2[m,4] > values2[m-1,4]):
                    m += 1
                    if m>=i:
                        break
                end = m
                if (values2[end,4] - values2[start,4])/values2[start,4] > U[i-60]:
                    #区间反弹超过阈值
                    flag = 1
                    break      
            else:
                m += 1
        if flag == 1:
            #说明区间里有反弹超过了
            continue
            #测试下一个j
        if flag2 == 0:
            #说明区间没有回弹的
            continue
            
        if i not in day_list and j not in fantan_qujian2:
            day_list.append(i)
            fantan_qujian.append(start_)
            fantan_qujian2.append(j)
```

```python
import matplotlib.ticker as ticker
sns.set(style='dark')
#plt.figure(figsize = (16,8),dpi = 300)
fig, ax = plt.subplots(1,1)
#ax.plot(x,s.IP)
fig.set_dpi(200)
ax.plot(values2[:,0],values2[:,4],linewidth = 1)
#ax.plot(day_list,values2[day_list,4],'*')
ax.xaxis.set_major_locator(ticker.MultipleLocator(70))
plt.xticks(rotation = 45)
plt.xlabel('Day')
plt.ylabel('Stock Price')
plt.title('Signal Day Visualization')
for i in range(len(day_list)):
    plt.axvline(day_list[i],ymax = (values2[day_list[i],4]-3200)/3000,alpha = 0.5,linewidth = 0.3 , color = 'r')

plt.savefig('statistic_day_0.jpg')
```

```python
time_sig = values2[day_list,0]
year_count = list(map(lambda x: x.split('-')[0],values2[day_list,0]))
plt.figure(dpi = 200)
plt.hist(year_count, align = 'left')

plt.xlabel('Year')
plt.ylabel('Toal Nums')
plt.title('Signal Day Counts')
plt.savefig('statistic_day.jpg')
```

```python
#计算每个信号日的所有股票的区间涨跌幅和前20日张跌幅。
values1 = data1.values[:,1:-2]
```

```python
get_group = {}
for i,day in enumerate(day_list):
    start = fantan_qujian[i]
    fudu1 = (values1[day] - values1[start]) / values1[start]
    fudu2 = 0
    for j in range(20):
        fudu2 += (values1[start-j] - values1[start-j-1]) / values1[start-j-1]
    fudu2 = fudu2/20
    final = fudu1 - fudu2
    argsort = np.argsort(final)[::-1]
    length = len(final)
    temp = []
    #for m in range(5):
    temp.append(argsort[:6])
    temp.append(argsort[6:12])
    temp.append(argsort[11:17])
    temp.append(argsort[16:22])
    temp.append(argsort[22:])
    
    get_group[day] = temp
```

```python
#开始买股票了
total_nums = 10000 #每支股的总价值为10000

money = {}
for i,day in enumerate(day_list[:-1]):
    indexs = get_group[day]
    temp = []
    nums = 10000/values1[day] #信号日每只股买的股数
    for index in indexs:
        start_price = values1[day,index]
        time = np.min((20,day_list[i+1]-day))
        end_price = values1[day+time,index]
        temp.append(np.sum((end_price - start_price)*nums[index]))
    money[day] = temp

day = day_list[-1]
indexs = get_group[day]
temp = []
nums = 10000/values1[day] #信号日每只股买的股数
for index in indexs:
    start_price = values1[day,index]
    time = np.min((20,1212-day))
    end_price = values1[day+time,index]
    temp.append(np.sum((end_price - start_price)*nums[index]))
money[day] = temp
```

```python
array_money = np.zeros((len(day_list),5))
for i in range(len(day_list)):
    array_money[i] = money[day_list[i]]
```

```python
total_money = np.sum(array_money,axis = 1)
total_money = np.cumsum(total_money)
```

```python
fig, ax = plt.subplots(1,1)

sns.set(style='darkgrid')
#ax.plot(x,s.IP)
fig.set_dpi(200)
plt.plot(time_sig,total_money)
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
plt.xticks(rotation = 45)
plt.title('Earned curve')
plt.savefig('earned_money.jpg')
```
